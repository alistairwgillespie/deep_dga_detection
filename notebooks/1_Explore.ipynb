# DGA Dataset - Explore

```python
# EDA Tools
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import dask.dataframe as dd

# Defaults
import os
import sys
```

```python
df = dd.read_csv('../data/train.csv', blocksize=64000000, names=['domain', 'dga_family'])
```

```python
dga_counts = df.groupby(['dga_family'])['domain'].agg('count')
```

```python
dga_counts.nlargest(20).compute().plot(kind='barh', stacked=False, figsize=[10,10], colormap='hsv')
plt.title('Top 20 DGA Families by Domains')
plt.ylabel('DGA Family')
plt.xlabel('Number of Domains')
plt.show()
```
